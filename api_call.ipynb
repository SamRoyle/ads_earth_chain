```python
import requests
import pprint

retrival_size = 1000
gsCertified = True

class API:
    def __init__(self, endpoint):
        self.endpoint = endpoint
        
    def get_data(self):
        response = requests.get(self.endpoint)
        return response.json()

project_api = API(f"https://api.goldstandard.org/projects?size={retrival_size}&gsCertified={gsCertified}")
gs_certified_project_list = project_api.get_data()

print(len(gs_certified_project_list))
```

```python
import pandas as pd

project_df = pd.DataFrame(gs_certified_project_list)
project_df.sort_values('id', inplace=True)
project_df.reset_index(drop=True,inplace=True)
project_df
```
